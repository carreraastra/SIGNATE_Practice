```python
import os
import cv2
import numpy as np
```

```python
def image_divide(image, size, overlap):
    
    '''
    <<画像分割関数>>
    <input>
    image　 入力画像(パス付)　　サイズは偶数　フォーマットbmp,jpg,..etc
    size 　 切り出す画像サイズ　サイズは偶数  縦横同じ
    overlap 重複領域のサイズ　　サイズは偶数
    
    <output> 
    分割後の画像　現状はjpg出力
    '''
    
    image_name = image.split('.')[0]
    img = cv2.imread(image)
   
    #print(img.shape[0])
    #print(img.shape[1])
    
    #縦サイズ　重複領域の有無に関わらず、切り出すサイズの倍数にするため、サイズを切り上げる
    if ((overlap == 0) and (img.shape[0] % size == 0)):#ジャストサイズは元画像のまま
        v_size = img.shape[0]
    else:
        if ( img.shape[0] % size <= (size - overlap) ):
            v_size = (img.shape[0] // size) * size + (size - overlap)
        else:
            v_size = (img.shape[0] // size + 1) * size + (size - overlap)
            
    #横サイズ　重複領域の有無に関わらず、切り出すサイズの倍数にするため、サイズを切り上げる
    if ((overlap == 0) and (img.shape[1] % size == 0)):#ジャストサイズは元画像のまま
        h_size = img.shape[1]
    else:
        if ( img.shape[1] % size <= (size - overlap) ):
            h_size = (img.shape[1] // size) * size + (size - overlap)
        else:
            h_size = (img.shape[1] // size + 1) * size + (size - overlap)
            
    #print(v_size)
    #print(h_size)

    #縦の分割数を求める
    if ((overlap == 0) and (img.shape[0] % size == 0)):
        v_split = (img.shape[0] // size)
    else:
        if ( img.shape[0] % size <= (size - overlap) ):
            v_split = (img.shape[0] // size) + 1
        else:
            v_split = (img.shape[0] // size) + 2
    
    #横の分割数を求める
    if ((overlap == 0) and (img.shape[1] % size == 0)):
        h_split = (img.shape[1] // size)
    else:
        if ( img.shape[1] % size <= (size - overlap) ):
            h_split = (img.shape[1] // size) + 1
        else:
            h_split = (img.shape[1] // size) + 2
            
    #print(v_split)
    #print(h_split)

    #サイズ増分を求める
    diff_v = int(v_size - img.shape[0])
    diff_h = int(h_size - img.shape[1])
    
    #print(diff_v)
    #print(diff_h)
    
    #サイズ増分は、0でパディングする
    img_pad = np.pad(img, [(0, diff_v), (0, diff_h), (0,0)], "constant")
    #print(img_pad.shape)
    #cv2.imwrite('img01_pad.jpg',img_pad)

    #画像分割（切り出しループ）
    for j, v in enumerate(np.arange(v_split)) :  
        for i, h in enumerate(np.arange(h_split)) :
            top = v * size + overlap 
            bottom = top + size
            left = h * size + overlap
            right = left + size
            cut_img = img_pad[top:bottom, left:right, :]
            
            #"./"にパスを入れる。フォーマットを変えたい場合は ".jpg"を変更する。
            cv2.imwrite(r"./" + r"\%s_%03.f_%03.f"%(image_name,i,j) + ".jpg",cut_img)
    return     
```
